```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cast=pd.read_csv('cast.csv')
cast.head()
```

```python
release_dates=pd.read_csv('release_dates.csv', index_col=2, parse_dates=['date'])
release_dates.head()
```

### Make a bar plot of the months in which movies with "Christmas" in their title tend to be released in the USA.

```python
rel = release_dates[(release_dates.index=='USA') & (release_dates['title'].str.contains("Christmas"))] 
rel.date.dt.month.value_counts().sort_index().plot(kind='bar')
```

### Make a bar plot of the months in which movies whose titles start with "The Hobbit" are released in the USA.

```python
rel = release_dates[(release_dates.index=='USA') & (release_dates['title'].str.startswith("Christmas"))] 
rel.date.dt.month.value_counts().sort_index().plot(kind='bar')
```

### Make a bar plot of the day of the week on which movies with "Romance" in their title tend to be released in the USA.

```python
rel = release_dates[(release_dates.index=='USA') & (release_dates['title'].str.contains("Romance"))] 
rel.date.dt.weekday.value_counts().sort_index().plot(kind='bar')
```

### Make a bar plot of the day of the week on which movies with "Action" in their title tend to be released in the USA.

```python
r = release_dates[(release_dates['title'].str.contains('Action')) & (release_dates.index=="USA")]
r.date.dt.weekday.value_counts().sort_index().plot(kind='bar')
plt.title('total weekday for movies having word "Action"')
plt.xlabel('total movies')
plt.ylabel('weekday')
```

### On which date was each Judi Dench movie from the 1990s released in the USA?

```python
u = release_dates[release_dates.index=='USA']
judi_90 = cast.loc[(cast['name']=='Judi Dench') & (cast['year']>=1990)]
judi_90.merge(u).sort_values('date')
```

### In which months do films with Judi Dench tend to be released in the USA?

```python
u = release_dates[release_dates.index=='USA']
judi = cast.loc[(cast['name']=='Judi Dench')]
judi_u = judi.merge(u).sort_values('date')

x = judi_u.date.dt.month.value_counts().sort_index()
x.plot(kind='bar')
plt.title("films release in months in which judi dench was in film")
plt.xlabel('month')
plt.ylabel('movie count')
```

### In which months do films with Tom Cruise tend to be released in the USA?

```python
u = release_dates[release_dates.index=='USA']
tom = cast.loc[(cast['name']=='Tom Cruise')]
tom_u = tom.merge(u).sort_values('date')

x = tom_u.date.dt.month.value_counts().sort_index()
x.plot(kind='bar')
plt.title("films release in months in which Tom Cruise was in film")
plt.xlabel('month')
plt.ylabel('movie count')
```
