## 3. Write a function called sed that takes as arguements a pattern string, a replacement string, and two filenames; it should read the first file and write the contents into the second file (creating it if necessary). If the pattern string appears anywhere in the file, it should be replaced with the replacement string. If an error occurs while opening, reading, writing, or closing files, your program should catch the exception, print an error message, and exit.

```python
# from file1, string will be replaced by rep_str and copies to file2
def sed(string, rep_str, file1, file2):
#    file_new = open(file2, "x")
#    file_new.close()
    try:
        with open(file1, "r") as f1, open(file2, "w") as f2:
            text = f1.read()
    
            text = text.replace(string, rep_str)
            f2.write(text)    
            
            f1.close()
            f2.close()
    except:
        print("Error is coming, check for errors")
    
    
sed('lion', 'elephant', 'story.txt', 'copy2.txt')
    
    
```

# 4. Log File Analysis:
## You have a log file containing records of user activities on a website. Each line in the file represents a log entry with details like timestamp, user ID, and action performed. Your task is to analyze this log file.   
## a) Write Python code to read the log file and extract specific information, such as the number of unique users or the most common action.
## b) How would you handle large log files efficiently without loading the entire file into memory?

```python
import json

flog = open("Android.log", "r")
lst = []
dd = ["date","time","machine_id","user_id"]

for line in flog:
    x = line.split()
    data = {k:v for k,v in zip(dd,x)}
    lst.append(data)
#print(lst)
    
for i in lst:
    print(json.dumps(i, indent = 4))
```
